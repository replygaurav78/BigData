##Python challenge question 10

http://www.pythonchallenge.com/pc/return/bull.html
There’s a picture of a bull and the clue “len(a[30]) = ?”. Clicking on the bull reveals “a = [1, 11, 21, 1211, 111221,
On careful observation,we see this is “look and say sequence”.




```python
def anyterm(term_string):
    out = ''
    previous = term_string[0]
    runlength = 1
    for i in term_string[1:]:
        if i == previous: # more of same
            runlength += 1
        else:         # run of new char; flush and reset
            out += str(runlength) + previous
            previous = i
            runlength = 1
    # final flush
    return out + str(runlength) + previous

term = '1'
for dmy in range(30):
    term = anyterm(term)
print len(term)


```

```python
import Image
 
img = Image.new( 'RGB', ( 100, 100 ) )
wire = Image.open( 'wire.png', 'r' )
 
xmin, xmax = 0, 99
ymin, ymax = 0, 99
 
pos = [ 0, 0 ]
place = 0
 
output = True
 
while not ( xmax - 1 == xmin or ymax - 1 == ymin ):
 
        for x in range( xmin, xmax + 1 ):
                if output: print "x ", pos[0], " y ", pos[1]
                pos[0] = x
                place += 1
                img.putpixel( pos, wire.getpixel( ( place, 0) ) )
        ymin += 1
        for y in range( ymin, ymax + 1 ):
                if output: print "x ", pos[0], " y ", pos[1]
                pos[1] = y
                place += 1
                img.putpixel( pos, wire.getpixel( ( place, 0) ) )
        xmax =xmax-1
        for x in range( xmax, xmin - 1, -1 ):
                if output: print "x ", pos[0], " y ", pos[1]
 
                pos[0] = x
                place += 1
                img.putpixel( pos, wire.getpixel( ( place, 0) ) )
        ymax = ymax-1
        for y in range( ymax, ymin - 1, -1 ):
                if output: print "x ", pos[0], " y ", pos[1]
                pos[1] = y
                place += 1
                img.putpixel( pos, wire.getpixel( ( place, 0) ) )
        xmin += 1
 
img.save( "spiral.png" )
```
