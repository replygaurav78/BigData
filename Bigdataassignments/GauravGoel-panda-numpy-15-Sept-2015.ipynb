```python
from numpy import random
```

```python
# standard normal distributed random numbers
C= random.randn(10,10)
C
```

```python
L=C.mean()
L
```

##Homework 2.3

2.3b Change all numbers under the mean to 0



```python
# create a zero matrix and substitute with elements from C matrix
data =zeros((10,10)) 
for i in range(10): 
        for j in range(10): 
            if C[i,j] < L: 
                data[i,j] = 0 
            else: 
                data[i,j] =C[i,j]
data
```

```python
Tdata=data.T
Tdata
```
